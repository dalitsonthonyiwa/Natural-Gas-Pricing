```python
import pandas as pd

def Pricing(inj_date, wit_date, inj_price, wit_price, inj_rate, wit_rate, max_storage, Storage_cost, data):
    """
  parameters
  inj_date: input "YYYY-MM-DD"
  wit_date: input "YYYY-MM-DD"
  inj_rate: input float the number of times it has been injected
  wit_rate: input float the number of times it has been withdrawn
  max_storage: input float
  Storage_cost: input float
  wit_price: input float the price of injection
  inj_price: input float the price of withdrawal
  data: input dataframe
    """

    # Convert input dates
    inj_date = pd.to_datetime(inj_date)
    wit_date = pd.to_datetime(wit_date)

    # Validate dates are in the price data
    if inj_date not in data.index or wit_date not in data.index:
        raise ValueError("Injection or withdrawal date not found in price data")

    # Get prices at injection and withdrawal dates
    unit_purchase_price = data.loc[inj_date, 'Prices']
    unit_sell_price = data.loc[wit_date, 'Prices']

    # Calculate storage days
    storage_days = (wit_date - inj_date).days
    if storage_days <= 0:
        raise ValueError("Withdrawal date must be after injection date")

    # Convert monthly storage cost to daily
    daily_storage_cost = Storage_cost / 30

    # Calculate quantities based on rates and storage period
    injected_quantity = inj_rate * max_storage  # Assuming 1 day injection
    withdrawn_quantity = max_storage * wit_rate  # Assuming 1 day withdrawal

    # Validate storage capacity
    if injected_quantity > max_storage:
        raise ValueError(f"Injected quantity {injected_quantity} exceeds max storage {max_storage}")

    # Calculate costs and revenues
    purchase_cost = unit_purchase_price * injected_quantity
    sales_revenue = unit_sell_price * withdrawn_quantity
    storage_cost = daily_storage_cost * storage_days
    total_inj_price = inj_price * inj_rate
    total_wit_price = wit_price * wit_rate

    # Calculate net value
    net_value = sales_revenue - (purchase_cost + storage_cost + total_inj_price + total_wit_price)

    return net_value
```
